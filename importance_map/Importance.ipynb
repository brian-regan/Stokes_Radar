```python
import scipy.ndimage
import numpy as np
import matplotlib.pyplot as plt
```

```python
pop = np.loadtxt('pop.mat')
precip = np.loadtxt('precip.mat')
```

```python
plt.imshow(pop)
plt.show()
```

```python
plt.imshow(precip)
plt.show()
```

```python
# Checking if my two grids match up
plt.imshow(precip + pop)
plt.show()
```

```python
# Rescale precip and pop to be 0-1 scale
pop_01 = pop/np.max(pop)
precip_01 = precip/np.max(precip)
```

```python
# Linear combination of importance factors:
# set these? - how important are precip and pop relatively
lambda_precip = 1
lambda_pop = 1
plt.imshow(precip_01 * lambda_precip + pop_01 * lambda_pop)
plt.show()
```
